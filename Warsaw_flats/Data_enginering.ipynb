## Balcony

```python
data.Balcony
```

```python
data.dropna(subset = ['Balcony'], axis = 'rows',inplace=True)
```

```python
data.Balcony.unique()
```

I will change this column into three new binary feature. Balkon, taras, ogródek.

```python
balkon = [1 if x == 'balkon, ogródek, taras' else 0 if  x =='balkon, taras' else 0 if x =='balkon' else 0   if x =='balkon, ogródek' else 0 for x in data.Balcony]
taras = [1 if x == 'balkon, ogródek, taras' else 0 if  x =='balkon, taras' else 0 if x =='taras' else 0   if x == 'ogródek, taras' else 0 for x in data.Balcony]
ogródek = [1 if x == 'balkon, ogródek, taras' else 0 if  x =='balkon, ogródek' else 0 if x =='ogródek' else 0   if x =='taras, ogródek' else 0 for x in data.Balcony]
```

```python
data['balkon'] = balkon
data['taras'] = taras
data['ogródek'] = ogródek
```

```python
data.head(4)
```

```python
data.drop(['Balcony'], axis = 1,inplace = True)
```
