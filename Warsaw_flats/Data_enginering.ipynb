## Price

### Too big values

```python
data[data['Price'] < 100000]
```

```python
price_z = (data.Price - data.Price.mean()) / data.Price.std()
print(len(price_z[price_z> 3]))
print(len(price_z[price_z< -3]))
```

```python
data.Price[price_z[price_z> 3].index].min()
```

```python
data.Price[price_z[price_z> 3].index].max()
```

```python
data.Price[price_z[price_z> 3].index].index
```

```python
data.drop(data.Price[price_z[price_z> 3].index].index, inplace= True)
```

```python
data.sort_values('Price',ascending=False, inplace=True)
data.head(60)
```

Biggest price values seems to be wrong. 55 milinos for 55 square meters is unreasonable. I'm going to drop those values. 

```python
big_values_index = data[data['Price']>= 28463659.0].index	
data.drop(big_values_index,inplace=True)
```
